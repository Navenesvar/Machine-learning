```python
import googletrans
import speech_recognition as sr
import pyaudio as pa
import gtts
import playsound
from translate import Translator
from IPython.display import Audio
from gtts import gTTS




recognizer=sr.Recognizer()
recognizer.energy_threshold = 20
translator=googletrans.Translator()
inputl='en'
outputl='ta'

try:
    with sr.Microphone() as source:
        print("Speak now")
        voice=recognizer.listen(source,timeout=20)
        text=recognizer.recognize_google(voice,language=inputl)
        print(text)
except:
    pass

translator = Translator(to_lang="ta")

translation3 = translator.translate(text)
print(translation3)
speech=gTTS(text=translation3,lang="ta",slow=False)
speech.save('output.mp3')
Audio('output.mp3')
```
